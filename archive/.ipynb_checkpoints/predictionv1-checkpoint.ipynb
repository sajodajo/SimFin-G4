```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12,5))
plot_acf(ts_prices.select('Differenced'), lags=40, ax=axes[0], color='red')
for patch in axes[0].collections:
    patch.set_facecolor('red')  # Change CI fill color
    patch.set_alpha(0.4)
plot_pacf(ts_prices.select('Differenced'), lags=40, ax=axes[1],color='green')
for patch in axes[1].collections:
    patch.set_facecolor('green')  # Change CI fill color
    patch.set_alpha(0.4)
axes[0].set_title("ACF Plot (Differenced)")
axes[1].set_title("PACF Plot (Differenced)")
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

additive_decomposition = seasonal_decompose(ts_prices.select('Close'), model="additive", period=12)
multiplicative_decomposition = seasonal_decompose(ts_prices.select('Close'), model="multiplicative", period=12)

residuals = multiplicative_decomposition.resid
trend = multiplicative_decomposition.trend
seasonal = multiplicative_decomposition.seasonal
```

```python
import itertools
from statsmodels.tsa.arima.model import ARIMA

diff_series = ts_prices.select('Differenced').drop_nulls().to_numpy().flatten()

# Define range of p, d, q values to test
p = range(0, 10)
d = [0] # Because it's an ARMA model, already stationary  
q = range(0, 10)

# Store results
results_list = []

# Loop through all possible combinations of p, d, q
for param in itertools.product(p, d, q):
    try:
        model = ARIMA(residuals, order=param)
        model_fit = model.fit()
        results_list.append([param, model_fit.aic, model_fit.bic])
    except:
        continue  # Skip invalid models

results_df = pl.DataFrame(
    {
        "order": [item[0] for item in results_list], 
        "AIC": [item[1] for item in results_list],   
        "BIC": [item[2] for item in results_list]  
    }
)

bestOrder = tuple(results_df.filter(pl.col("AIC") == results_df["AIC"].min()).select('order').row(0)[0])
```

```python
model = ARIMA(diff_series, order=bestOrder)
arma_results = model.fit()
```

```python


import matplotlib.pyplot as plt

ts_prices = ts_prices.with_columns(pl.col("Date").cast(pl.Date).dt.year().alias("Year"))


train = ts_prices.filter(pl.col("Year") < 2024)
test = ts_prices.filter(pl.col("Year") == 2024)
train_plot = ts_prices.filter(pl.col("Year") == 2023)

in_sample_forecast = arma_results.fittedvalues[:train.shape[0]]


# Convert Polars Date column to NumPy array for plotting
train_dates = train["Date"].to_numpy().flatten()
train_close = train["Close"].to_numpy().flatten()


final_arma_forecast = in_sample_forecast * trend * seasonal

plt.figure(figsize=(12, 6))
plt.plot(train_dates, train_close, label="Actual Data", marker='o', linestyle="-", color="black")
plt.plot(train_dates, final_arma_forecast, label="In-Sample Forecast", marker='o', linestyle="--", color='red')

plt.xlabel("Date")
plt.ylabel("Index")
plt.title("ARMA In-Sample Forecast vs Actual Data")
plt.legend()
plt.grid()
plt.show()
```
